```python
import pandas as pd
```

# Selecting Queue-ID from Offline Feasible Sheet

```python
main_df1 = pd.read_excel('Data/main.xlsx', sheet_name = "Offline Feasible")
main_df1
```

```python
df1 = pd.DataFrame()
df1['Queue'] = main_df1['Queue'].copy()
# Removing NaN values
df1.dropna(inplace = True)
df1.head(50)
```

```python
# Removing 'Queue -ID' rows from the column 'Queue-ID'
df1 = df1[df1['Queue'] != 'Queue-ID']
df1.head(50)
```

# Selecting Queue-ID from Online Feasible Sheet

```python
main_df2 = pd.read_excel('Data/main.xlsx', sheet_name = "Online Feasible")
main_df2
```

```python
df2 = pd.DataFrame()
df2['Queue'] = main_df2['Queue'].copy()
# Removing NaN values
df2.dropna(inplace = True)
df2.head(50)
```

```python
# Removing 'Queue -ID' rows from the column 'Queue-ID'
df2 = df2[df2['Queue'] != 'Queue-ID']
df2.head(50)
```

# Selecting Queue-ID from Online Feasible (2) Sheet

```python
main_df3 = pd.read_excel('Data/main.xlsx', sheet_name = "Online Feasible (2)")
main_df3
```

```python
df3 = pd.DataFrame()
df3['Queue'] = main_df3['Queue'].copy()
# Removing NaN values
df3.dropna(inplace = True)
df3.head(50)
```

```python
# Removing 'Queue -ID' rows from the column 'Queue-ID'
df3 = df3[df3['Queue'] != 'Queue-ID']
df3.head(50)
```

# Selecting Queue-ID from Online Not Feasible  Sheet

```python
main_df4 = pd.read_excel('Data/main.xlsx', sheet_name = "Online Not Feasible")
main_df4
```

```python
df4 = pd.DataFrame()
df4['Queue'] = main_df4['Queue'].copy()
# Removing NaN values
df4.dropna(inplace = True)
df4.head(50)
```

```python
# Removing 'Queue -ID' rows from the column 'Queue-ID'
df4 = df4[df4['Queue'] != 'Queue-ID']
df4.head(50)
```

```python
# Create a list of all the dfs
dfs = [df1, df2, df3, df4]

# Concatenate them row-wise (vertically)
merged_df = pd.concat(dfs, ignore_index=True)

# Reset the index of the merged DataFrame
merged_df.reset_index(drop=True, inplace=True)

merged_df.head(50)
```

```python
merged_df.shape
```

```python
# Remove duplicate rows based on all columns
merged_df.drop_duplicates(inplace=True)

# Reset the index of the DataFrame
merged_df.reset_index(drop=True, inplace=True)
merged_df.head(10)
```

```python
merged_df.shape
```

# Importing database dataset to compare merged_df

```python
database = pd.read_csv('Data/Database.csv')
database.head(50)
```

```python
database['Match'].fillna(0, inplace=True)
database
```

```python
# Create a set of unique 'Queue' values from 'merged_df' for faster lookup
merged_df_values = set(merged_df['Queue'])

# Use a lambda function with apply to check if each 'Queue' value in 'database' is in 'merged_df'
database['Match'] = database['Queue'].apply(lambda x: 1 if x in merged_df_values else 0)
```

```python
database.head(50)
```

```python
database['Match'].value_counts()
```

```python
filt = database['Match'] == 0
output1 = database[filt]
output1
```

```python
# saving the final output
database.to_excel('Data/output.xlsx')
output1.to_excel('Data/output1.xlsx')
```
