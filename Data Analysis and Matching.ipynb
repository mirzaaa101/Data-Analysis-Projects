Generating OLS Regression Report

```python
import statsmodels.api as sm
```

```python
X = merged_df_sorted1[['Patent_Public', 'Bill']]
y = merged_df_sorted1['Stock_Price']

# Add a constant column to the input features (OLS model requires an intercept term)
X = sm.add_constant(X)

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression results
print(model.summary())
```

```python
OLS_Regression1 = model.summary().as_text()
```
