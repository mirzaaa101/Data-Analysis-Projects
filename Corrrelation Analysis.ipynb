Dataset Preparation

```python
import numpy as pd
import pandas as pd
```

```python
# DATA IMPORT
df_sales = pd.read_excel('Data/Correlation_testing.xlsx', sheet_name = 'Sales')
df_sales
```

```python
# Renaming the unnamed column to date
df_sales.rename(columns={'Unnamed: 0': 'date'}, inplace=True)
df_sales
```

```python
# Only date, Product_A1 and Product A_2 will need to be comapre. So. I will keep the those three column only
df_sales = df_sales[['date','Product_A1','Product_A2']]
df_sales
```

```python
# Change columne names of sheet sales
df_sales.rename(columns={'Product_A1':'Sales_A1', 'Product_A2':'Sales_A2'}, inplace=True)
df_sales
```

Same Procedure for Volumns Sheet

```python
df_volumes = pd.read_excel('Data/Correlation_testing.xlsx', sheet_name = 'Volumes')
df_volumes
```

```python
df_volumes.rename(columns={'Unnamed: 0':'date','Product_A1':'Volumes_A1', 'Product_A2':'Volumes_A2'}, inplace=True)
df_volumes
```

```python
df_volumes = df_volumes[['date','Volumes_A1','Volumes_A2']]
df_volumes
```

```python
# Concatenating two dfs
merged_df = pd.merge(df_sales, df_volumes, on='date', how='inner')
merged_df
```

```python
merged_df['year'] = pd.to_datetime(merged_df['date']).dt.year
# Reorder the columns
merged_df = merged_df[['date', 'year'] + [col for col in merged_df.columns if col not in ['date', 'year']]]
merged_df
```
