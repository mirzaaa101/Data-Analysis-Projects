```python
import pandas as pd
```

```python
offline_feasible = pd.read_excel("Data/Forcast_Automation_forecast_v5.xlsx", sheet_name = "Offline Feasible")
offline_feasible
```

```python
of_queue = pd.DataFrame()
of_queue['Queue'] = offline_feasible['Queue'].copy()
# Removing NaN values
of_queue.dropna(inplace = True)
# Removing 'Queue -ID' rows from the column 'Queue-ID'
of_queue = of_queue[of_queue['Queue'] != 'Queue-ID']
of_queue.head(10)
```

```python
Intent_Online_Today = pd.read_excel("Data/Forcast_Automation_forecast_v5.xlsx", sheet_name = "Intent Online Today")
Intent_Online_Today
```

```python
iot_queue = pd.DataFrame()
iot_queue['Queue'] = Intent_Online_Today['Queue'].copy()
# Removing NaN values
iot_queue.dropna(inplace = True)
# Removing 'Queue -ID' rows from the column 'Queue-ID'
iot_queue = iot_queue[iot_queue['Queue'] != 'Queue-ID']
iot_queue.head(10)
```

```python
Online_Feasible = pd.read_excel("Data/Forcast_Automation_forecast_v5.xlsx", sheet_name = "Online Feasible")
Online_Feasible
```

```python
onf_queue = pd.DataFrame()
onf_queue['Queue'] = Online_Feasible['Queue'].copy()
# Removing NaN values
onf_queue.dropna(inplace = True)
# Removing 'Queue -ID' rows from the column 'Queue-ID'
onf_queue = onf_queue[onf_queue['Queue'] != 'Queue-ID']
onf_queue.head(10)
```

```python
Online_Not_Feasible = pd.read_excel("Data/Forcast_Automation_forecast_v5.xlsx", sheet_name = "Online Not Feasible")
Online_Not_Feasible
```

```python
onnf_queue = pd.DataFrame()
onnf_queue['Queue'] = Online_Not_Feasible['Queue'].copy()
# Removing NaN values
onnf_queue.dropna(inplace = True)
# Removing 'Queue -ID' rows from the column 'Queue-ID'
onnf_queue = onnf_queue[onnf_queue['Queue'] != 'Queue-ID']
onnf_queue.head(10)
```

```python
# Create a list of all the dfs
dfs = [of_queue, iot_queue, onf_queue, onnf_queue]

# Concatenate them row-wise (vertically)
merged_df = pd.concat(dfs, ignore_index=True)

# Reset the index of the merged DataFrame
merged_df.reset_index(drop=True, inplace=True)

merged_df.head(10)
```

```python
merged_df.shape
```

```python
Database = pd.read_excel("Data/Forcast_Automation_forecast_v5.xlsx", sheet_name = "Database")
Database
```

```python
Database['Match'] = 0
Database
```

```python
merged_df_values = set(merged_df['Queue'])
Database['Match'] = Database['Queue'].apply(lambda x: 1 if x in merged_df_values else 0)
```

```python
Database['Match'].value_counts()
```

```python
filt = (Database['Match'] == 0)
queue_names = Database[filt]
queue_names
```
